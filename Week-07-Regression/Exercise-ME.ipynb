# Rough Guide a v1 model.
0. Make a model to predict PTS 
0. For v1, dont use `2P, 3P, FG, or FT`.  Instead use `2PA, 3PA, FTA`.  
0. Include any other cols use see fit.  


````python
# Calculate total points
df['TotalPoints'] = df['2P'] * 2 + df['3P'] * 3 + df['FT']

# Calculate average total points for teams that made the playoffs
avg_playoff_points = df[df['Playoffs'] == 1]['TotalPoints'].mean()

# Set the figure size for better visibility
plt.figure(figsize=(10, 8))

# Bar plot of TotalPoints vs Team, differentiated by Playoffs
sns.barplot(x='TotalPoints', y='Team', data=df, hue='Playoffs')

# Add a horizontal line for the average points required to make playoffs
plt.axvline(avg_playoff_points, color='r', linestyle='--',label='Avg Points to Make Playoffs')

plt.title('Total Points vs Team with Playoff Qualification')
plt.xlabel('Total Points')
plt.ylabel('Team')
plt.legend()
plt.show()
````
